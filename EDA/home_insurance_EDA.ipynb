# Home Insurance EDA

This notebook contains some of the work on EDA for the home insurance dataset.

```python
import numpy as np
import pandas as pd
from pandas_profiling import ProfileReport
import random

random.seed(42)
```

```python
df = pd.read_csv("../Data/Raw_data/home_insurance.csv")
df.head(10)
```

```python
# Use pandas profiling to quickly generate an initial report

df = df.sample(frac=0.01, random_state=1)
profile = ProfileReport(df, title="Home Insurance Data Report")
profile.to_file("home_insurance_report.html")
```
